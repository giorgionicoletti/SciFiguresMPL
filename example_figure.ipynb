```python
import matplotlib
import matplotlib.pyplot as plt

import numpy as np
```

```python
# generate some data
np.random.seed(42)

x1 = np.linspace(0, 1, 100)
y1a = np.exp(x1)
y2a = np.exp(2*x1)
y3a = np.exp(3*x1)

x2 = np.random.randn(100)
y2 = np.random.rand(100)

x3 = np.linspace(0, 2*np.pi, 100)
y3 = np.random.rand(100)

x4 = np.linspace(0, 2*np.pi, 100)
y4a = np.sin(x4)
y4b = np.cos(x4)

y5 = np.random.gamma(3, size=1000)
```

```python
fig, ax = plt.subplot_mosaic([["pan1", "pan1", "pan2", "pan3", "pan3"],
                              ["pan1", "pan1", "pan2", "pan3", "pan3"],
                              ["pan4", "pan4", "pan4", "pan5", "pan5"]], figsize=(10, 4))

plt.subplots_adjust(wspace=0.6, hspace=0.8)

plt.show()
```

```python
data_x = [x1, x2, x3, x4]
data_y = [[y1a, y2a, y3a], [y2], [y3], [y4a, y4b]]
```

```python
fig, ax = plt.subplot_mosaic([["pan1", "pan1", "pan2", "pan3", "pan3"],
                              ["pan1", "pan1", "pan2", "pan3", "pan3"],
                              ["pan4", "pan4", "pan4", "pan5", "pan5"]], figsize=(10, 4))

plt.subplots_adjust(wspace=0.6, hspace=0.8)

for i in range(1, 5):
    ax["pan" + str(i)].set_xlabel("$x_" + str(i) + "$")
    ax["pan" + str(i)].set_ylabel("$y_" + str(i) + "$")
    for idx, y in enumerate(data_y[i-1]):
        if i == 2:
            ax["pan" + str(i)].scatter(data_x[i-1], y)
        else:
            ax["pan" + str(i)].plot(data_x[i-1], y, label = "Label " + str(idx + 1))

ax["pan5"].hist(y5, density = True)
ax["pan5"].set_xlabel("data")
ax["pan5"].set_ylabel("pdf")

ax["pan1"].legend()

plt.show()
```
